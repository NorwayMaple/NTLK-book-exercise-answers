```python
%matplotlib inline

import nltk
import matplotlib.pyplot as plt
import numpy
```

```python
# NTLK book Chapter 2 Exercise 23a
def zipf(text):
    fdist = nltk.FreqDist(w.lower() for w in text)

    # Would it be wiser to use numpy arrays instead of lists?
    fdist_sorted = fdist.most_common(1000)
    freq = [datum[1]/sum(fdist.values()) for datum in fdist_sorted]
    plt.plot([rank * freq for rank, freq in enumerate(freq, start = 1)])
    #The x axis should start at 1 but instead starts at 0
    plt.title('Zipf\'s Law k')
    plt.show()
    '''
Each diagonal represents an integer number of appearances of the word.  At the beginning, stop words seem not to obey
constant k because Zipf's law may not apply to stopwords
'''

# Try including different works of literature on same plot
```

```python
# Chapter 2 Problem 24

# See problem desription for meaning of n

# Careful, this function should be ablet to take cfdist as a FreqDist or a list of FreqDists
def generate_model(cfdist, word, n=3, num=200):
    words = []
    if type(cfdist) == type(nltk.probability.ConditionalFreqDist()):
        for i in range(num):
            words.append(word)
            word = random.choice(cfdist[word].most_common(n))[0]
    elif type(cfdist) == type(list()) and len(cfdist) > 1:
        # Cannot sum list of ConditionalFreqDist's because you can't initialize an empty one to start the summation
        # Instead we initialize it with the first addend and add the rest sequentially
        combined_cfdist = cfdist[0]
        for i in range(1, len(cfdist)):
            combined_cfdist += cfdist[i]
        for i in range(num):
            words.append(word)
            word = random.choice(combined_cfdist[word].most_common(n))[0]
    else:
        raise Exception('Invalid cfdist')
            
            
    return words

text = nltk.corpus.inaugural.words('2001-Bush.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
text2 = nltk.corpus.gutenberg.words('shakespeare-caesar.txt')
bigrams2 = nltk.bigrams(text2)
cfd2 = nltk.ConditionalFreqDist(bigrams2)
cfd_list = [cfd, cfd2]


```

```python
print(' '.join(generate_model(cfd, 'America')))
```

```python
# This primarily uses Shakespeare's vocabular because the Shakespeare text is longer.
# Also, specific content words like Clinton and Brutus are less likely to be used as opposed to common English words
print(' '.join(generate_model(cfd_list, 'America')))
```

```python
# Chapter 2 Exercise 25

def find_languages(word):
    
    langs = []
    fileids = nltk.corpus.udhr.fileids()
    for fileid in fileids:
        if fileid[-6:] != 'Latin1':
            continue
        words = nltk.corpus.udhr.words(fileid)
        if word in words:
            lang = fileid[:-7]
            langs.append(lang)
    return langs
```
