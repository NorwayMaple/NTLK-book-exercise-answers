```python
%matplotlib inline

import nltk
import matplotlib.pyplot as plt
import numpy
```

```python
# NTLK book Chapter 2 Exercise 23a
def zipf(text):
    fdist = nltk.FreqDist(w.lower() for w in text)

    # Would it be wiser to use numpy arrays instead of lists?
    fdist_sorted = fdist.most_common(1000)
    freq = [datum[1]/sum(fdist.values()) for datum in fdist_sorted]
    plt.plot([rank * freq for rank, freq in enumerate(freq, start = 1)])
    #The x axis should start at 1 but instead starts at 0
    plt.title('Zipf\'s Law k')
    plt.show()
    '''
Each diagonal represents an integer number of appearances of the word.  At the beginning, stop words seem not to obey
constant k because Zipf's law may not apply to stopwords
'''

# Try including different works of literature on same plot
```

```python
# Chapter 2 Problem 24

# See problem desription for meaning of n

# Careful, this function should be ablet to take cfdist as a FreqDist or a list of FreqDists
def generate_model(cfdist, word, n=3, num=200):
    words = []
    if type(cfdist) == type(nltk.probability.ConditionalFreqDist()):
        for i in range(num):
            words.append(word)
            word = random.choice(cfdist[word].most_common(n))[0]
    elif type(cfdist) == type(list()) and len(cfdist) > 1:
        # Cannot sum list of ConditionalFreqDist's because you can't initialize an empty one to start the summation
        # Instead we initialize it with the first addend and add the rest sequentially
        combined_cfdist = cfdist[0]
        for i in range(1, len(cfdist)):
            combined_cfdist += cfdist[i]
        for i in range(num):
            words.append(word)
            word = random.choice(combined_cfdist[word].most_common(n))[0]
    else:
        raise Exception('Invalid cfdist')
            
            
    return words

text = nltk.corpus.inaugural.words('2001-Bush.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
text2 = nltk.corpus.gutenberg.words('shakespeare-caesar.txt')
bigrams2 = nltk.bigrams(text2)
cfd2 = nltk.ConditionalFreqDist(bigrams2)
cfd_list = [cfd, cfd2]


```

```python
print(' '.join(generate_model(cfd, 'America')))
```

```python
# This primarily uses Shakespeare's vocabular because the Shakespeare text is longer.
# Also, specific content words like Clinton and Brutus are less likely to be used as opposed to common English words
print(' '.join(generate_model(cfd_list, 'America')))
```

```python
# Chapter 2 Exercise 25

def find_languages(word):
    
    langs = []
    fileids = nltk.corpus.udhr.fileids()
    for fileid in fileids:
        if fileid[-6:] != 'Latin1':
            continue
        words = nltk.corpus.udhr.words(fileid)
        if word in words:
            lang = fileid[:-7]
            langs.append(lang)
    return langs
```

```python
find_languages('por')
```

```python
# Chapter 2 Exercise 26

# This is for English hyponyms

import itertools

tot_hyponyms = 0
n = 0

synsets = nltk.corpus.wordnet.all_synsets('n')

# For troubleshooting:
# synsets = itertools.islice(synsets, 10000)

for synset in synsets:
    length = len(synset.hyponyms())
    if length == 0:
        continue
    else:
        tot_hyponyms += length
        n += 1
        
ave_hyponyms = tot_hyponyms/n
print(ave_hyponyms)
```

```python
# Chapter 2 Exercise 26

from nltk.corpus import wordnet as wn
from collections import Counter

# list of pos's
poses = [wn.NOUN, wn.VERB, wn.ADJ, wn.ADV]

synsets = wn.all_synsets('n')

# For troubleshooting:
# synsets = itertools.islice(synsets, 10000)

for pos in poses:
    count = Counter()
    # Each synset is a meaning
    for synset in wn.all_synsets(pos = pos):
        for lemma in synset.lemmas():
            count[str(lemma.name())] += 1
    print('Polysemy for ' + str(pos))
    print(sum(count.values())/len(count))
```

```python
# Chapter 2 Exercise 28
pairs = [('car', 'automobile'), ('gem', 'jewel'), ('journey', 'voyage'), ('boy', 'lad'), ('coast', 'shore'), ('asylum', 'madhouse'), ('magician', 'wizard'), ('midday', 'noon'), ('furnace', 'stove'), ('food', 'fruit'), ('bird', 'cock'), ('bird', 'crane'), ('tool', 'implement'), ('brother', 'monk'), ('lad', 'brother'), ('crane', 'implement'), ('journey', 'car'), ('monk', 'oracle'), ('cemetery', 'woodland'), ('food', 'rooster'), ('coast', 'hill'), ('forest', 'graveyard'), ('shore', 'woodland'), ('monk', 'slave'), ('coast', 'forest'), ('lad', 'wizard'), ('chord', 'smile'), ('glass', 'magician'), ('rooster', 'voyage'), ('noon', 'string')]



for pair in pairs:
    print(pair)
    # Synset pairs: all the pairwise combinations of each synset for the words in pair
    synset_pairs = [(synset1, synset2) for synset1 in wn.synsets(pair[0]) for synset2 in wn.synsets(pair[1])] # This needs to be fixed - can I used list comprehension here?
    #similarities: The similarities for all these pairwise combination, including Nones
    similarities = [synset1.path_similarity(synset2) for synset1, synset2 in synset_pairs] 
    #similarities now excludes Nones
    similarities = [similarities for similarities in similarities if similarities is not None]
    similarity = max(similarities)
    print(similarity)
```

```python
# Chapter 3 Exercise 39

# If you want to allow apostrophes see the str.upper() workaround using regex at
# https://docs.python.org/3.6/library/stdtypes.html#str.title

import re

# Converts letters to numbers that correspond to those letters
def letters_to_numbers(s):
    s = re.sub(r'[bfpv]', r'1', s)
    s = re.sub(r'[cgjkqsxz]', r'2', s)
    s = re.sub(r'[dt]', r'3', s)
    s = re.sub(r'l', r'4', s)
    s = re.sub(r'[mn]', r'5', s)
    s = re.sub(r'r', r'6', s)
    return s
    

def soundex(s):
    lo = s.lower()
    
    #Step 1
    first_char = lo[0]
    other_chars = lo[1:]
    
    #Step 2
    a = letters_to_numbers(other_chars)
    
    # Step 3
    b = letters_to_numbers(first_char) + a
    # h and w are removed here in b and c required in Step 1
    c = re.sub(r'(.)[hw]?\1', r'\1', b)
    d = re.sub(r'[aeiouyhw]', r'', c[1:])
    e = first_char + d
    # Step 4
    f = e
    while len(f) < 4:
        f += '0'
    result = f[:4].upper()
    print(s + ": Original word")
    print(lo + ": Lower case")
    print(other_chars + ': Remove first char')
    print(a + ": Substitute letters to numbers where possible for non-first chars")
    print(b + ": Do the same with the first letter and prepend it")
    print(c + ": Remove duplicate letters when adjacent or separated by h or w")
    print(d + ": Remove first letter (in case it is a vowel), and then remove a, e, i, o, u, y, h, w, which have no numbers")
    print(e + ": Put first char back, which you saved at the beginning")
    print(f + ': Add the 0\'s')
    print(result + ': Truncate to 4 chars and make upper case')
    return result

print(soundex('Robert'))
print()
print(soundex('Rupert'))
print()
print(soundex('Rubin'))
print()
print(soundex('Ashcroft'))
print()
print(soundex('Ashcraft'))
print()
print(soundex('Tymczak'))
print()
print(soundex('Pfister'))
print()
print(soundex('Honeyman'))
```
